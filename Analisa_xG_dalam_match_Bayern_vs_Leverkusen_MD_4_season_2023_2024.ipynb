```python
import pandas as pd
from statsbombpy import sb

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
events = sb.events (match_id = 3895074)
df = events
df.info()
```

```python
events = sb.events (match_id = 3895074)
df = events[events.type == 'Shot']
df = df[['period', 'minute', 'shot_statsbomb_xg', 'team', 'player', 'shot_outcome' ]]
df.rename(columns = {'shot_statsbomb_xg':'xG', 'shot_outcome':'result'}, inplace = True)
df.sort_values(by='team', inplace=True)
df.head()
```

```python
hteam = df['team'].iloc[0]
ateam = df['team'].iloc[-1]
print('Home Team : ' + hteam)
print('Away Team : ' + ateam);
```

```python
# Cumulative Sum xG home team
h_df = df[df['team'] == hteam].copy()  # Buat salinan eksplisit
h_df.sort_values(by='minute', inplace=True)
h_df['h_cum'] = h_df['xG'].cumsum()

# Cumulative Sum xG away team
a_df = df[df['team'] == ateam].copy()  # Buat salinan eksplisit
a_df.sort_values(by='minute', inplace=True)
a_df['a_cum'] = a_df['xG'].cumsum()

pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.width', 1000)  # Mengatur lebar tampilan

# Menampilkan DataFrame h_df dan a_df
print(h_df)
print(a_df)
```

```python
# Filter untuk hasil gol dari tim home
h_goal = h_df[h_df['result'].str.contains("Goal")].copy()  # Buat salinan eksplisit
h_goal['scorechart'] = h_goal['minute'].astype(str) + "'" + " " + h_goal['player']

# Filter untuk hasil gol dari tim away
a_goal = a_df[a_df['result'].str.contains("Goal")].copy()  # Buat salinan eksplisit
a_goal['scorechart'] = a_goal['minute'].astype(str) + "'" + " " + a_goal['player']

# Menampilkan hasil
print("Home Team Goals:")
print(h_goal[['minute', 'player', 'scorechart']])

print("Away Team Goals:")
print(a_goal[['minute', 'player', 'scorechart']])
```

```python
#Total xG
a_total = round(a_df['xG'].sum(),2).astype(str)
h_total = round(h_df['xG'].sum(),2).astype(str)
```

Visualisasi

```python
import matplotlib.pyplot as plt

# Buat figure dan axis
fig, ax = plt.subplots(figsize=(15, 6))

# Total xG
h_total = h_df['h_cum'].iloc[-1]  # Total xG untuk tim home
a_total = a_df['a_cum'].iloc[-1]  # Total xG untuk tim away

# Buat plot garis
ax.step(x=h_df['minute'], y=h_df['h_cum'], where='post', color='#004d98', label=f'{hteam} ({h_total:.2f} xG)', linewidth=2)
ax.step(x=a_df['minute'], y=a_df['a_cum'], where='post', color='red', label=f'{ateam} ({a_total:.2f} xG)', linewidth=2)

# Buat scatter plot untuk menandai gol
ax.scatter(x=h_goal['minute'], y=h_goal['h_cum'], marker='o', s=200, color='#004d98', zorder=5)
ax.scatter(x=a_goal['minute'], y=a_goal['a_cum'], marker='o', s=200, color='red', zorder=5)

# Kustomisasi chart
for j, txt in enumerate(h_goal['scorechart']):
    ax.annotate(txt, (h_goal['minute'].iloc[j], h_goal['h_cum'].iloc[j]), xycoords='data', ha='center',
                xytext=(-50, 10), textcoords='offset points', arrowprops=dict(arrowstyle="->", color='white'))

for i, txt in enumerate(a_goal['scorechart']):
    ax.annotate(txt, (a_goal['minute'].iloc[i], a_goal['a_cum'].iloc[i]), xycoords='data', ha='center',
                xytext=(-50, 10), textcoords='offset points', arrowprops=dict(arrowstyle="->", color='white'))

# Mengatur ticks dan grid
plt.xticks([0, 15, 30, 45, 60, 75, 90])
plt.yticks([0, 0.5, 1, 1.5, 2, 2.5, 3])
plt.grid()

# Menambahkan teks
fig.text(s=f'{hteam} - {ateam} (2 - 2)', x=0.125, y=0.97, fontsize=18, fontweight="bold")
fig.text(s='Bundesliga 23/24', x=0.125, y=0.92, fontsize=12)

# Menambahkan legenda
legend = ax.legend(title='Total Expected Goals (xG)', loc='best', shadow=True)
legend.get_frame().set_facecolor('lightgrey')  # Ganti warna latar belakang legenda

# Menambahkan label sumbu
plt.ylabel("Expected Goals (xG)", fontsize=12, labelpad=20)
plt.xlabel("Minutes", fontsize=12, labelpad=20)

# Menampilkan plot
plt.show()
```

Berdasarkan analisa diatas gol pertama tercipta dari kubu bayern, gol diciptakan oleh harry kane pada menit ke 6 dengan xG diatas 0.5, selanjutnya dibalas dari kubu leverkusen dengan gol dari alex grimaldo pada menit ke 23 dengan xG dibawah 0.5, lalu kubu bayern menciptakan gol melalui leon pada menit 85 dengan xG diatas 2 dan dibalas juga dari kubu leverkusen dengan gol dari alejandro palacios menit 93 dengan xG diatas 2
